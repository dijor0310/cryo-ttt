```python
from monai.data import SlidingPatchWSIDataset
import pandas as pd
```

```python
tomo_list = pd.read_csv("./src/configs/tomo_list.csv")
tomo_paths = tomo_list['tomo_path'].to_list()
gt_paths = tomo_list['gt_path'].to_list()
data = [{"image": tomo, "label": gt} for tomo, gt in zip(tomo_paths, gt_paths)]
```

```python
dataset = SlidingPatchWSIDataset(data=data, patch_size=64, overlap=0.0, include_label=True)
```

```python
from monai.transforms import RandSpatialCropSamples
from monai.data import PatchDataset
sampler = sampler = RandSpatialCropSamples(roi_size=64, num_samples=5,
                                 random_center=True, random_size=False)


ds = PatchDataset(data=data,
                  patch_func=sampler,
                  samples_per_image=5,)
```
